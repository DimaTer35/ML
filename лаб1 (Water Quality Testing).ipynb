```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Загрузка данных
data = pd.read_csv('D:/ТМО/Water Quality Testing.csv')
```

```python
# Размер набора данных
dataset_size = data.shape
print("Размер набора данных:", dataset_size)
```

```python
# Список столбцов
columns = data.columns.tolist()
print("Список столбцов:", columns)
```

```python
# Первые 5 строк данных
head = data.head()
print("Первые 5 строк данных:")
print(head)
```

```python
# Сводная статистика
summary = data.describe()
print("Сводная статистика:")
print(summary)
```

```python
# Визуализация распределения характеристик воды
sns.pairplot(data)
plt.show()
```

```python
# Гистограмма распределения значений турбидности
plt.figure(figsize=(8, 6))
sns.histplot(data['Turbidity (NTU)'], bins=20, kde=True, color='steelblue')
plt.xlabel('Турбидность (NTU)')
plt.ylabel('Частота')
plt.title('Распределение значений турбидности')
plt.show()
```

```python
# Точечная диаграмма: температура vs. уровень растворенного кислорода
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Temperature (°C)', y='Dissolved Oxygen (mg/L)', data=data, color='steelblue')
plt.xlabel('Температура (°C)')
plt.ylabel('Уровень растворенного кислорода (мг/л)')
plt.title('Связь между температурой и уровнем растворенного кислорода')
plt.show()
```

```python
# Матрица корреляции
correlation_matrix = data.corr()
```

```python
# Визуализация матрицы корреляции
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Матрица корреляции признаков')
plt.show()
```
